# GRÁFICOS DE HABILIDADES.

### Pasos previos.

Librerías.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Carga de datos.

```python
path_dataset = '..\\datosTransformados\\scrapping_es\\'
file_name = 'skill.csv'
file_name1='soft_skill.csv'

skills=pd.read_csv(path_dataset+file_name, index_col=[0]) 
soft_skills = pd.read_csv(path_dataset+file_name1, index_col=[0]) 
```

### Gráficos.

Gráfico de barras con importancia de habilidades técnicas.

```python
fig1,ax1=plt.subplots()
ax1.bar(skills['habilidad'],skills['conteo'])
ax1.set_title('Conteo de habilidades técnicas')
ax1.set_xlabel('Palabras')
ax1.set_ylabel('Conteo')
fig1.tight_layout()
fig1.set_size_inches(15,5)
plt.xticks(rotation=50)
fig1.savefig('../gráficos/conteo_habilidades_tecnicas.png')
```

Gráfico de barras con conteo de habilidades personales.

```python
fig2,ax2=plt.subplots()
ax2.bar(soft_skills['habilidad'],soft_skills['conteo'])
ax2.set_title('Conteo de habilidades técnicas')
ax2.set_xlabel('Palabras')
ax2.set_ylabel('Conteo')
fig2.tight_layout()
fig2.set_size_inches(15,5)
plt.xticks(rotation=50)
fig2.savefig('../gráficos/conteo_habilidades_personales.png')
```

Gráfico de tarta con importancia de habilidades técnicas importantes.

```python
skills_importantes=skills[skills['conteo']>=2]
```

```python
fig3,ax3=plt.subplots()
ax3.pie(skills_importantes['importancia %'],labels=skills_importantes['habilidad'])
ax3.set_title('Importancia de habilidades técnicas relevantes')
fig3.tight_layout()
fig3.savefig('../gráficos/tarta_habilidades_tec.png')
```

Gráfico de tarta con importancia de habilidades personales importantes.

```python
soft_skills_importantes=soft_skills[soft_skills['conteo']>=2]
```

```python
fig4,ax4=plt.subplots()
ax4.pie(soft_skills_importantes['importancia %'],labels=soft_skills_importantes['habilidad'])
ax4.set_title('Importancia de habilidades personales relevantes')
fig4.tight_layout()
fig4.savefig('../gráficos/tarta_habilidades_per.png')
```
